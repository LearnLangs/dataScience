```python
def getHead():
    from matplotlib import pyplot as plt
    import seaborn as sb
    
    # now change the figure size
    plt.rcParams['figure.figsize'] = [8,2] 
    
    # set grid and import data
    sb.set_style('darkgrid')
    
    titanic = sb.load_dataset('titanic')
    # display the first five - or head
    return titanic.head()
    

getHead()
```

```python
# importing the necessary libraries
from matplotlib import pyplot as plot
import seaborn as sb

# now we have to resize the figure
plot.rcParams["figure.figsize"] = [8,2]

# setting the grid for display
sb.set_style("darkgrid")

# importing the titanic dataset 
Titanic = sb.load_dataset("titanic")

# calling the Titanic variable
Titanic.head()
```

```python
sb.countplot(x="pclass", data=Titanic)
```

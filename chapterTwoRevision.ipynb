```python
# importing necessay libs
from matplotlib import pyplot as PLOT
import seaborn as SB

# resize the figure
PLOT.rcParams["figure.figsize"] = [8,2]

# set grids
SB.set_style("darkgrid")

#load dataset
titanicData = SB.load_dataset("titanic")

titanicData.head()

```

```python
SB.countplot(x="pclass", data=titanicData)

```

```python
SB.distplot(titanicData['fare'], kde=True)
```

```python
SB.countplot(x="survived", data=titanicData)
```

```python
SB.barplot(x="class", y="fare", data=titanicData)
```

```python
SB.barplot(x="embark_town", y="age", data=titanicData, color="red")
```

```python
class tryOOP:
    """
        This OOP is just meant for example purposes and illustrate my understanding in OOP
    """
    name = "davisone"
    age = 19
    smart = True
    
    def check(self):
        return(f"His name is {self.name}, he is {self.age} years, and is he smart ? {self.smart}")
        

h = tryOOP()
print(h.check())   
```

```python
class nextOOP:
    name = "davisone"
    age = 19
    smart = False
    
    def __init__(self, name = name, age = age, smart = smart):
        nextOOP.name = name
        nextOOP.age = age
        nextOOP.smart = smart
        
    def check(self):
        return(f"Her name is {self.name}, she is {self.age}, and is she smart? {self.smart}")
    
b = nextOOP("Babe", 20, True)
print(b.check())
```

```python
titanic = titanic[['survived', 'pclass', 'age', 'fare']]
titanic.head()
```

```python
titanic.isnull() # is used to check for null values
```

```python
# now we have to find the mean values
titanic.isnull().mean() # applying mean on the null values to insert the at the missing spaces
```

# Interpretation of the Output above
    the output above simply signifies that, only the age filed has missing values, thus around 19.87% of the in the age field is missing

```python
# Finding the mean of the none missing values in age column
mean = titanic.age.mean()
print(mean)
```

```python
#finding the median
median = titanic.age.median()
print(median)
```

```python
# creating a new column in the data set for mean age and median age
titanic["mean"] = titanic.age.fillna(mean)
titanic["median"] = titanic.age.fillna(median)

titanic.head(20)
```

```python
import numpy as np

titanic['mean'] = np.round(titanic['mean'], 1)

titanic.head(20)
```

```python
plt.rcParams['figure.figsize'] = [8,6]

fig = plt.figure()

aux = fig.add_subplot(111)

titanic['age'].plot(kind='kde', ax=aux)
titanic['median'].plot(kind='kde', ax=aux, color='yellow')
titanic['mean'].plot(kind='kde', ax=aux, color='green')

lines, lables = aux.get_legend_handles_labels()

aux.legend(lines, lables, loc='best')
```
